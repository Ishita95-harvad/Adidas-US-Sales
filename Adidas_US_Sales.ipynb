## Data loading

### Subtask:
Load the "Adidas US Sales.csv" file into a Pandas DataFrame.


**Reasoning**:
Load the "Adidas US Sales.csv" file into a pandas DataFrame and display the first 5 rows.



```python
import pandas as pd

try:
    df = pd.read_csv('Adidas US Sales.csv')
    display(df.head())
except FileNotFoundError:
    print("Error: 'Adidas US Sales.csv' not found. Please ensure the file exists in the current directory.")
    df = None # Assign None to df in case of error
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None
```

## Data exploration

### Subtask:
Explore the loaded Adidas sales data to understand its characteristics.


**Reasoning**:
I need to explore the data by examining its shape, info, descriptive statistics, histograms for numerical features, missing values, and unique values in categorical columns.



```python
# Data Shape and Info
print("Data Shape:", df.shape)
print("\nData Info:")
df.info()

# Descriptive Statistics
print("\nDescriptive Statistics:")
numerical_cols = ['total_sales', 'units_sold', 'operating_profit', 'price_per_unit']
display(df[numerical_cols].describe())

# Data Distribution (Histograms)
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 8))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Missing Values Analysis
print("\nMissing Values Analysis:")
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
display(missing_df)

# Unique Values
print("\nUnique Values in Categorical Columns:")
categorical_cols = ['retailer', 'region', 'state', 'city', 'product', 'sales_method']
for col in categorical_cols:
    print(f"\nUnique values for {col}:")
    print(df[col].unique())
```

## Data cleaning

### Subtask:
Clean the Adidas sales data by identifying and handling outliers in the 'total_sales', 'units_sold', and 'operating_profit' columns.


**Reasoning**:
I will detect and handle outliers in 'total_sales', 'units_sold', and 'operating_profit' using the IQR method, then generate box plots to verify the changes.



```python
import matplotlib.pyplot as plt

# Columns to handle outliers
cols_to_handle = ['total_sales', 'units_sold', 'operating_profit']

for col in cols_to_handle:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Replace outliers with bounds
    df[col] = df[col].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))

# Box plots for verification
plt.figure(figsize=(12, 6))
for i, col in enumerate(cols_to_handle):
  plt.subplot(1, 3, i + 1)
  plt.boxplot(df[col])
  plt.title(f'Box Plot of {col}')
  plt.ylabel(col)
plt.tight_layout()
plt.show()
```

## Data wrangling

### Subtask:
Create new features and convert data types as needed within the `df` DataFrame.


**Reasoning**:
Create new features, convert data types, and handle potential errors as specified in the instructions.



```python
import pandas as pd

# 1. Calculate Profit Margin
df['profit_margin'] = df.apply(lambda row: (row['operating_profit'] / row['total_sales']) * 100 if row['total_sales'] != 0 else 0, axis=1)

# 2. Average Price Per Unit (Data type verification and conversion if necessary)
if not pd.api.types.is_numeric_dtype(df['price_per_unit']):
    try:
        df['price_per_unit'] = pd.to_numeric(df['price_per_unit'], errors='coerce')
    except Exception as e:
        print(f"Error converting 'price_per_unit' to numeric: {e}")
        # Handle the error appropriately (e.g., fill with the mean or median)
        # For this example, fill with the mean
        df['price_per_unit'] = df['price_per_unit'].fillna(df['price_per_unit'].mean())


# 3. Convert Invoice Date
try:
    df['invoice_date'] = pd.to_datetime(df['invoice_date'], format='%Y-%m-%d', errors='coerce')
except ValueError as e:
    print(f"Error converting 'invoice_date': {e}")
    # Handle invalid date formats (e.g., replace with NaT or a default date)
    # Example: Replace invalid dates with NaT
    df['invoice_date'] = pd.to_datetime(df['invoice_date'], errors='coerce')

# 4. (Optional) Feature Engineering based on Product Category
# Example: Extract product category (assuming product names include categories)
df['product_category'] = df['product'].str.extract(r'(.*)[\s-]')
display(df.head())
```

## Data analysis

### Subtask:
Analyze sales trends over time, identify top-performing products, and analyze relationships between sales, units sold, retail price, and other relevant variables.


**Reasoning**:
Analyze sales trends over time, identify top-performing products, and analyze relationships between sales, units sold, retail price, and other relevant variables. Group by invoice date, calculate monthly and quarterly sales aggregates, identify top performers based on total sales and units sold, and analyze relationships between key variables.



```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Sales Trends Over Time
df['invoice_date'] = pd.to_datetime(df['invoice_date'])
df = df.set_index('invoice_date')
monthly_sales = df.resample('M')[['total_sales', 'units_sold']].sum()
quarterly_sales = df.resample('Q')[['total_sales', 'units_sold']].sum()

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['total_sales'], label='Monthly Total Sales')
plt.plot(quarterly_sales['total_sales'], label='Quarterly Total Sales')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Sales Trend Over Time')
plt.legend()
plt.show()


# 2. Top-Performing Products
top_products_sales = df.groupby('product')['total_sales'].sum().sort_values(ascending=False).head(10)
top_products_units = df.groupby('product')['units_sold'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.bar(top_products_sales.index, top_products_sales.values, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Top Products by Total Sales')
plt.subplot(1,2,2)
plt.bar(top_products_units.index, top_products_units.values, color='lightcoral')
plt.xticks(rotation=45, ha='right')
plt.title('Top Products by Units Sold')
plt.tight_layout()
plt.show()


# 3. Relationship Analysis
correlation_matrix = df[['total_sales', 'units_sold', 'price_per_unit', 'profit_margin']].corr()
display(correlation_matrix)

plt.figure(figsize=(10, 8))
plt.scatter(df['total_sales'], df['units_sold'], color='green', alpha=0.5)
plt.xlabel('Total Sales')
plt.ylabel('Units Sold')
plt.title('Relationship between Total Sales and Units Sold')
plt.show()


# 4. Regional/State Performance
regional_sales = df.groupby('region')['total_sales'].sum().sort_values(ascending=False)
state_sales = df.groupby('state')['total_sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.bar(regional_sales.index, regional_sales.values, color='orange')
plt.title('Regional Sales Performance')
plt.subplot(1,2,2)
plt.bar(state_sales.index, state_sales.values, color='purple')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 States by Sales')
plt.tight_layout()
plt.show()
```

## Data visualization

### Subtask:
Visualize the key findings from the data analysis. Create visualizations to represent sales trends, product performance, and relationships between variables.


**Reasoning**:
Visualize the key findings from the data analysis as requested in the instructions, using appropriate chart types and visual elements.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Sales Trends Over Time
plt.figure(figsize=(12, 6))
plt.plot(df.resample('M')['total_sales'].sum(), label='Monthly Total Sales', color='skyblue')
plt.plot(df.resample('Q')['total_sales'].sum(), label='Quarterly Total Sales', color='coral')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Sales Trend Over Time')
plt.legend()
plt.show()

# 2. Top-Performing Products
top_10_products_sales = df.groupby('product')['total_sales'].sum().nlargest(10)
top_10_products_units = df.groupby('product')['units_sold'].sum().nlargest(10)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
top_10_products_sales.plot(kind='barh', ax=axes[0], color='skyblue')
axes[0].set_title('Top 10 Products by Total Sales')
axes[0].set_xlabel('Total Sales')
top_10_products_units.plot(kind='barh', ax=axes[1], color='lightcoral')
axes[1].set_title('Top 10 Products by Units Sold')
axes[1].set_xlabel('Units Sold')
plt.tight_layout()
plt.show()

# 3. Relationships between Variables
correlation_matrix = df[['total_sales', 'units_sold', 'price_per_unit', 'profit_margin']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(df['total_sales'], df['units_sold'], alpha=0.5, color='green')
plt.xlabel('Total Sales')
plt.ylabel('Units Sold')
plt.title('Total Sales vs. Units Sold')
plt.show()

# 4. Regional and State Performance
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
df.groupby('region')['total_sales'].sum().plot(kind='bar', color='orange')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')

plt.subplot(1, 2, 2)
df.groupby('state')['total_sales'].sum().nlargest(10).plot(kind='bar', color='purple')
plt.title('Top 10 States by Total Sales')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
